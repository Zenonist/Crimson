```python
import pickle
import numpy as np
import pandas as pd
```

```python
id_list = ['018_087_123','022_094','041_050_095','042_072_169','058_093_144']
filename = id_list[0]
df_base = pd.read_csv('../Data/Junior_Data/' + filename + '.csv')
df_base
```

```python
ground_truth = df_base.iloc[:,4:].values
df_base.drop(df_base.columns[[4,5,6,7,8,9,10,11]],axis = 1, inplace= True)
print(ground_truth)
```

```python
df_base
```

```python
pickle_prediction = np.load('../WangChanBERT/pickle_files/Prob_Predict_' + filename + '.pickle', allow_pickle=True)
pickle_prediction = np.round(pickle_prediction)
pickle_prediction = pickle_prediction.astype(int)
```

```python
print(pickle_prediction)
```

```python
resulttype = []
for i in range(len(pickle_prediction)):
    counter = 0
    for j in range(8):
        if (pickle_prediction[i][j] != ground_truth[i][j]):
            counter = counter + 1
    if (counter == 0):
        resulttype.append('Correct')
    else:
        if (ground_truth[i].tolist().count(1) == 0):
            resulttype.append('GT_AllZeroes')
        elif (ground_truth[i].tolist().count(1) < pickle_prediction[i].tolist().count(1)):
            resulttype.append('MisType1')
        elif (ground_truth[i].tolist().count(1) > pickle_prediction[i].tolist().count(1)):
            resulttype.append('MisType2')
        else:
            resulttype.append('MisType3')
        
```

```python
ground_truth_str_arr = []
for i in range(len(ground_truth)):
    ground_truth_str  = ' '.join([str(x) for x in ground_truth[i]])
    ground_truth_str_arr.append(ground_truth_str)
```

```python
prediction_str_arr = []
for i in range(len(pickle_prediction)):
    prediction_str  = ' '.join([str(x) for x in pickle_prediction[i]])
    prediction_str_arr.append(prediction_str)
```

```python
b = pd.DataFrame(ground_truth_str_arr,columns= ['Ground Truth'])
c = pd.DataFrame(prediction_str_arr,columns= ['Prediction'])
d = pd.DataFrame(resulttype,columns= ['ResultType'])
```

```python
df_result = pd.concat([df_base,b,c,d],axis=1)
```

```python
df_result
```

```python
df_result = df_result[df_result.ResultType != 'Correct']
```

```python
df_result.to_csv('./WangChanBERT/Error_Analysis_'+filename+'.csv',encoding = 'utf-8')
```
